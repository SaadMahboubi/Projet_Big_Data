# Post Cryptopanic 

```python
import pprint as pp

      
print(db_Crypto_Panic.list_collection_names())
for postcrpp in postCryptoPanic.find():
    pp.pprint(postcrpp)
```

```python
postCryptoPanic=postcrpp['post']
```

```python
import pandas as pd

df= pd.DataFrame(postCryptoPanic)  

# on separe les différents éléments contenus dans 'votes' et on ajoute les colonnes 'cryptos'
df= pd.concat([df.drop(['votes'], axis=1), df['votes'].apply(pd.Series)], axis=1)

#On separe egalement les différents éléments contenus dans 'currencies'
dfc= pd.json_normalize(df['currencies'])
# Elément 1
dfc1=pd.json_normalize(dfc[0])
dfc1= dfc1['code']
# Elément 2 
dfc2=pd.json_normalize(dfc[1])
dfc2= dfc2['code']
# Elément 3 
dfc3=pd.json_normalize(dfc[2])
dfc3= dfc3['code']
# Elément 4
dfc4=pd.json_normalize(dfc[3])
dfc4= dfc4['code']
#On concatene  ces 4 élémens
cryptos= pd.concat([dfc1, dfc2, dfc3, dfc4], axis=1)
# on renomme les colonnes
cryptos.columns=['crypto1', 'crypto2', 'crypto3', 'crypto4']

# On concatene  'df' et 'cryptos' en supprimant la colonne 'currencies'
df= pd.concat([df.drop(['currencies'], axis=1), cryptos], axis =1)
#On renome 'published at' par 'date'
df.rename(columns={'published_at': 'dateheure'}, inplace=True)

#On reorganise les colonnes
df= df[['id', "dateheure",'title','crypto1', 'crypto2', 'crypto3', 'crypto4','negative','positive', 'important',	'liked', 'disliked','lol','toxic',	'saved','comments','url']]

df['dateheure'] = pd.to_datetime(df['dateheure'])  
df
```
